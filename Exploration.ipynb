## Production well 3: Test results

```python
# import the packages we need in this exercise
import numpy as np
from math import pi, e
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

font = {'family': 'serif',
        'color':  'green',
        'weight': 'bold',
        'size': 16,
        }

day_to_second = 86400

# Additional measured data
B   = 1.0         # volume factor, V_res/V_surf changes with bubble point
mu  = 3.664e-4          # viscosity, unit: Pa/s https://www.engineeringtoolbox.com/water-dynamic-kinematic-viscosity-d_596.html?vA=353.8&units=K#
# comp= 2e-10      # compressibility , 1/Pa depends on phase, storativity, etc. 
h   = 140          # thickness, unit: m
q   = 200 / day_to_second       # rate, unit: m3/d to m3/s
rw  = 0.2159        # well radius, unit: m

# # convert unit to SI unit
# mu  = 8.9e-4 #viscosity
# q   = ??

# load pressure and time data from the .txt file and plot pressure vs time
pressure = np.genfromtxt('P03_welltest.txt', skip_header = 5, usecols = 2 )#pressure in bars must be converted to Pascals
pressure = pressure * 100000
time = np.genfromtxt('P03_welltest.txt', skip_header = 5, usecols = 0) * day_to_second
plt.figure(figsize=(15, 7))
plt.plot(time, pressure,'g')
plt.title('pressure vs. time', fontdict=font)
ax = plt.gca()
plt.xlabel('time / s', fontdict=font)
ax.xaxis.set_minor_formatter(mticker.ScalarFormatter())
ax.ticklabel_format(style='plain', axis='x')#getting rid of scientific notation
plt.ylabel('pressure / Pa', fontdict=font)
plt.show()
```

```python
for i in range(len(pressure)):
    if pressure[i+1] > pressure[i]: #next pressure should be smaller than last. if not then break
        break
pressure_drop = pressure[:i+1] #record point of pressure build up
ind = i

plt.figure(figsize=(15, 7))
plt.semilogx(time[:ind+1], pressure_drop,'r')
plt.title('pressure vs. log(time)', fontdict=font)
plt.xlabel('log(time / s)', fontdict=font)
plt.ylabel('pressure / Pa', fontdict=font)
plt.show()
```

### The line is pretty straight in the production well. We can select any area along the straight length as transient flow.

```python
for i in range(len(time)):
    if time[i] > 500:
        break
ind_start = i 

for i in range(len(time)):
    if time[i] > 80000:
        break
ind_end = i

pressure_wellbore = pressure_drop[:ind_start]
pressure_transient = pressure_drop[ind_start:ind_end]
pressure_late = pressure_drop[ind_end:]

# transient flow 
P = pressure_transient
T = time[ind_start:ind_end]

num_coef = 2
num_points = len(P)

# Now regress data to line

A = np.ones((num_points, num_coef))#make the matrix
for ii in range(num_points):
    '''coefficient at logarithm c'''
    A[ii, 0] = 1
    A[ii, 1] = np.log(T[ii])
    
rhs = np.zeros((num_points))#numpy array, right hand side
rhs[:] = P

solution = np.linalg.solve(np.dot(A.T, A), np.dot(A.T, rhs))#left hand side A transposed by A, etc. ; solution for least square method

pressure_fit = np.zeros((num_points))
for ii in range(num_points):
    pressure_fit[ii] = solution[0] + solution[1] * np.log(T[ii])

# visualize the results
plt.figure(figsize=(15, 7))
plt.semilogx(time[:ind_start], pressure_wellbore,'g*-',time[ind_start:ind_end], pressure_transient,'bd-',
            time[ind_end:ind+1], pressure_late,'ko-')
plt.semilogx(T,pressure_fit,'r')
plt.legend(['Wellbore storage effect','Transient flow','Late trasient flow','Linear regression'],fontsize = 15)
plt.title('pressure vs. time', fontdict=font)
plt.xlabel('time / s', fontdict=font)
plt.ylabel('pressure / Pa', fontdict=font)
plt.show()

# calculate permeability from pw = -mlnt + const

k = -1e15 * q * B * mu / (4 * np.pi * h * solution[1])

print("Permeability of Production Well 03 = "+str(k)+" mDarcy")
```

## Production Well 03 lifetime and pump energy

```python
# 1 PJ = 1e12 kJ
import pandas as pd
from math import pi
import numpy as np
h = 140 #reservoir height
w = 800 #assumed width of reservoir
k = 224e-15 #mD permeability (-12:1000s)(-13:100s)(-14:10s) (-15:1s)
year = 365 #days in year
phi = 0.205 #porosity
 
temp_inj = 20 #injection temperature in celsius
rho_w = 975 # kg/m3 density of water
rho_r = 2700 # kg/m3 density of rock
C_w = 4.1   # kJ/kg/C heat capacity of water ((enthalpy of water at T2 - enthalpy of water at T1)/T2-T1)
C_r = 0.8532  # kJ/kg/C heat capacity of rock; sandstone: 0.715 + 0.00170T + 1.908 + ((10^-6)*T^2) where T is reservoir temp

mu  = 3.664e-4          # viscosity, unit: Pa/s https://www.engineeringtoolbox.com/water-dynamic-kinematic-viscosity-d_596.html?vA=353.8&units=K#
r_0 = 0.2159   # m well radius

nu = 0.4 #pump efficiency; geothermal direct use utilization rate is usually 30-45%, Reykjavik is 51%

day_to_second = 86400

T_ini = 77.135 # or you can calculate based on the thermal gradient, initial temperature in celsius

L = [400,600,800,1000]  # well distance

Q = 0.1e12 / 365 / (rho_w * C_w * (T_ini - temp_inj)) #m3/day

dT = T_ini - temp_inj

thermal_power = (Q*0.01157) * C_w * dT

A = w * h #width * height

v = rho_w * C_w * Q / (rho_r * C_r * (1-phi) * A) #m/day

lt_list = []
WI_list = []
dp_list = []
EP_list = []
for i in L:
#     print(i)
    lt = i / v / 365
#     print(lt)
    lt_list.append(lt)
#     print(lt_list)
    wi = 2 * np.pi * k * h / (np.log( i /r_0) * mu)
    WI_list.append(wi)
    dp = Q / wi
    dp_list.append(dp)
    Ep = Q * dp / nu #pump energy 
    EP_list.append(Ep)
#         return lt_list, WI_list, dp_list, EP_list
# make_lists(L, v, mu, r_0, Q)
# print(lt_list)

# WI = 2 * np.pi * k * h /(np.log(L/r_0) * mu) #well index
# # print(r_0)
# dp = Q / WI #differential pressure

# Ep = Q * dp / nu #pump energy 

# print('   Q = ' + str(Q) + ' m3/day')
# print('   Velocity = ' + str(v) + ' m/year')
# print('   Lifetime = ' + str(lt) + ' years')
# print('Pump energy = ' + str(Ep*1e-15/365) + ' PJ')

# import csv

dataset_variable = [
    ['Well separation meters', 'Flow rate m3/day', 'Reservoir lifetime years', 'pump energy PJ', 'differential temp C', 'power MW', 'velocity m/year','height', 'width', 'area', 'initial temperature', 'well radius'],
    [L[0], Q, lt_list[0], EP_list[0]*1e-15/365, dT, thermal_power/1000, v, h, w, A, T_ini, r_0],
    [L[1], Q, lt_list[1], EP_list[1]*1e-15/365, dT, thermal_power/1000, v, h, w, A, T_ini, r_0],
    [L[2], Q, lt_list[2], EP_list[2]*1e-15/365, dT, thermal_power/1000, v, h, w, A, T_ini, r_0],
    [L[3], Q, lt_list[3], EP_list[3]*1e-15/365, dT, thermal_power/1000, v, h, w, A, T_ini, r_0]
]
# print(dataset_variable)
# # well_.csv gets created in the current working directory
# with open ('well_prod03.csv','w',newline = '') as csvfile:
#     my_writer = csv.writer(csvfile, delimiter = ' ')
#     my_writer.writerows(dataset_variable)
df_1 = pd.DataFrame(dataset_variable, columns=dataset_variable[0])
df_1 = df_1.iloc[1: , :]
df_1.head()
```
